## PART 1

#### First we load the data

```python
import numpy as np
import pylab as plt
f1 = 'asteroids5000.csv'

data = np.genfromtxt(f1,dtype='float',delimiter=',')
```

#### Then we identify the "nan" values in each column

```python
nanvals = sum(np.isnan(data))

for i in range(len(nanvals)):
    print 'There are %i nan values in column %i' % (nanvals[i],i+1)
```

#### Next we quantify the number of complete rows

```python
datacheck = sum(np.isnan(data.transpose()))
goodrows = np.where(datacheck == 0)

print '\nThe number of complete rows is %i\n' % len(goodrows[0])
```

#### And then we create a new array of only good values and compute some basic statistics

```python
gooddata = data[goodrows]
cols = gooddata.transpose()

for i in range(len(cols)):
    print 'For column %i-- max: %f\t| min: %f\t\t| mean: %f\t| stdev: %f' % (i+1,np.max(cols[i]),np.min(cols[i]),np.mean(cols[i]),np.std(cols[i]))
```

## PART 2

#### Let's create a scatter plot of the semi-major axis vs the sine of the inclination angle

```python
%pylab inline

semimaj,ecc,inc,H,diam,BV,UB = cols
inc = inc*(np.pi/180.)
plt.scatter(semimaj,np.sin(inc),s=1,c='k',edgecolor='none')
plt.xlim(2.0,4.5),plt.ylim(0,0.5)
plt.xlabel('Semi-major Axis (a)'),plt.ylabel('sin(i)')
```

###### There's so many gaps in the data!  Are these simply incomplete measurements (as any missing value would discount a row of data in its entirety)? Or is this saying something about different resonance bands in the asteroid belt?  We need more data!

#### Let's create a scatter plot of the absolute magnitude vs the B-V color

```python
plt.scatter(BV,H,s=1,c='k',edgecolor='none')
plt.plot([0.79,0.79],[0,20],ls='--',c='k')
plt.xlim(0.5,1.1),plt.ylim(14,4)
plt.xlabel('B-V'),plt.ylabel('Absolute Mag (H)')
```

```python
red = np.where(BV > 0.79)
blue = np.where(BV < 0.79)
plt.scatter(semimaj[red],np.sin(inc[red]),s=1,c='r',edgecolor='none')
plt.scatter(semimaj[blue],np.sin(inc[blue]),s=1,c='b',edgecolor='none')
plt.xlim(2.0,4.5),plt.ylim(0,0.5)
plt.xlabel('Semi-major Axis (a)'),plt.ylabel('sin(i)')
```

###### Even though there's only a small amount of data being worked with (< 800 objects), there does appear to be a trend where the redder objects generally have smaller orbits than bluer objects.  Comparison to the paper shows that those gaps in semi-major axis are not simply from incomplete measurements or any selection bias, but are actually the Kirkwood gaps.
